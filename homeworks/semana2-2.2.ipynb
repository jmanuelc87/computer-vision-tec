```python
import matplotlib.pyplot as plt
import numpy as np
import base64
import cv2
```

```python
image1 = cv2.imread('../datasets/image1.bmp')
image2 = cv2.imread('../datasets/image2.bmp')

image3 = cv2.subtract(image1, image2)

image4 = image3 + 109

images = [image1, image2, image3, image4]
labels = ['Original', 'Background', 'Low Brightness', 'Result']

fig, axes = plt.subplots(1, 4, figsize=(15,9))
fig.tight_layout(pad=2.7)

axes = axes.ravel()

for i in range(4):
    axes[i].set_title(labels[i])
    axes[i].imshow(images[i])
```
